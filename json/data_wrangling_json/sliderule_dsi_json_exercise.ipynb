## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
#Opening/ Loading the JSON file as String
with open('data/world_bank_projects.json') as data_file:
    jfile = json.load(data_file)
```

```python
#Load the file as pandas dataframe
df = json_normalize(jfile)
```

```python
#Viewing the first 5 entries of the dataframe
df.head()
```

```python
# Question 1. Find the 10 countries with most project

# Getting a dataframe with the countryname and projects
df_country = df[['countryname', 'project_name']]

#View the loaded df
df_country.head()
```

```python
#Printing the top 10 values from the df_country
df_country['countryname'].value_counts().head(10)
```

```python
# Question 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

# As seen from the dataframe df head O/P initially, the column "mjtheme_namecode" contains nested JSON elements.
# We need to normalize the same.

#Normalising the JSON file on the 'mjtheme_namecode' column
project_themes = json_normalize(data=jfile, record_path='mjtheme_namecode')

#View the normalised df
project_themes.head()

```

```python
#Printing the top 10 themes names
project_themes['name'].value_counts().head(10)
```

```python
# Question 3. In 2. above you will notice that some entries have only the code and the name is missing. 
#           Create a dataframe with the missing names filled in.

# Replace the empty strings with "Not Available"
project_themes['name'].replace(r'^\s*$', 'Not Available', regex=True, inplace=True)

#Print the top 10 themes names
project_themes['name'].value_counts().head(10)
```
